```python
import pandas as pd
df_synergy = pd.read_csv("synergy_logistics_database.csv", index_col=0, encoding='utf-8', parse_dates=[4, 5])
df_synergy
```

#Rutas mas demandadas

```python
#Synergy logistics está considerando la posibilidad de enfocar sus esfuerzos en las 10 rutas más demandadas, segun las exp y las imp
df_rutas= df_synergy.groupby(by=["origin", "destination"])
df_rutas = df_rutas.describe()["total_value"]
df_rutas_demandadas=df_rutas["count"]
df_rutas_demandadas= df_rutas_demandadas.sort_values(ascending=False).head(10)
print(df_rutas_demandadas)

df_rutas_valor= df_synergy.groupby(by=["origin", "destination"]).sum().sort_values(by="total_value",ascending=False).head (10)
print("Las 10 rutas con mayor valor total son: \n")
print(df_rutas_valor)
```

#Transporte

```python
#¿Cuáles son los 3 medios de transporte más importantes para Synergy logistics considerando el valor de las 
#importaciones y exportaciones? ¿Cuál es medio de transporte que podrían reducir? 
df_transporte= df_synergy.groupby(by=["transport_mode"]).sum().sort_values(by="total_value",ascending=False)
print(df_transporte)
```

#Valor Total

```python
#gpo de paises que le generan el 80% de sus ingresos

exports = df_synergy[df_synergy['direction'] == 'Exports']
imports = df_synergy[df_synergy['direction'] == 'Imports']

pais_total_value = exports.groupby('origin').sum()['total_value'].reset_index()
total_value_for_percent = pais_total_value['total_value'].sum()
pais_total_value['percent'] = 100 * pais_total_value['total_value'] / total_value_for_percent
pais_total_value.sort_values(by='percent', ascending=False, inplace=True)
pais_total_value['cumsum'] = pais_total_value['percent'].cumsum()

lista_pequena = pais_total_value[pais_total_value['cumsum'] < 85]
print(lista_pequena)


pais_total_value = imports.groupby('origin').sum()['total_value'].reset_index()
total_value_for_percent = pais_total_value['total_value'].sum()
pais_total_value['percent'] = 100 * pais_total_value['total_value'] / total_value_for_percent
pais_total_value.sort_values(by='percent', ascending=False, inplace=True)
pais_total_value['cumsum'] = pais_total_value['percent'].cumsum()

lista_pequena = pais_total_value[pais_total_value['cumsum'] < 85]
print(lista_pequena)
```
